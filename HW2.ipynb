## Задание №2 "Фильтр"

Напишите программу, которая проводит первичную обработку неких сложных и
глючных логов. Нужно удалить сочетание «%%» в начале некоторых строк и удалить
строки, начинающиеся с «####».


**Пример:**

**ввод:**

3

SVO TRS 29481292

%%LJPZ DME 11113283675

###&%^^^^

**вывод:**

SVO TRS 29481292

LJPZ DME 11113283675


```python
n = int(input('Введите кол-во строк'))
rows = [input() for i in range(n)]
print('Результат обработки фильтра:')
for row in rows:
    if row[:4] != '####':
        if row[:2] == '%%':
            print(row[2:])
        else:
            print(row)
        
```

## Задание №3 "От и до"

Напишите программу, которая считывает список чисел, а затем выводит сумму в
диапазоне от заданного до заданного. При этом в программе не должно
использоваться обращение к элементам по индексу.

На первой строке вводится натуральное число N — количество чисел
в списке. 

Далее следуют N целых чисел из списка, каждое на отдельной
строке. 

Затем на отдельных строках следуют начальный (p) и конечный (q) номера
элементов списка, который нужно просуммировать. Нумерация идёт с единицы. Гарантируется,
что 1≤ p ≤ q ≤ N.


**Пример:**

**Ввод:**

5

100

222

300

444

500

2

4

**Вывод:**

966

**↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓Важно↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓↓**

**Замечание №1**

Эта задача решается просто через срез, т.е. sum(arr[p-1:q]). Но здесь мы ссылаемся на индексы

**Замечание №2**

Здесь использовался метод pop, так как если он не на, что не ссылается, то удаляет последний элемент массива (но на выходе дает выкинутый элемент, который нам не нужен). Методы del требует ссылку на индексы (даже в выражении del(arr[0]) мы ссылаемся на 0 индекс. Так что идея заключалась в том, чтобы выкинуть нужное кол-во элементов с конца, затем записать оставшийся список в обратном порядке, используя метод reverse, а затем вновь использовать нужное кол-во раз метод pop. Опять же всю эту длинную кухню с мешаниной методов можно было избежать, просуммировав срез: sum(arr[p-1:q])
Скорость преложенного варианта O(p) + O(q) + O(n) + O(q-p+1), для среза O(q - p + 1), что указывает на то, что срез в разы эффективней

**↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑Важно↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑↑**

```python
n = int(input('Введите кол-во чисел'))
arr = [int(input()) for i in range(n)]
p = int(input('Введите начальный номер элемента списка, который нужно просуммировать: '))
q = int(input('Введите конечный номер элемента списка, который нужно просуммировать: '))

for i in range(n-q):
    arr.pop()
arr.reverse()
for i in range(p-1):
    arr.pop()
print(sum(arr))
```

## Задание №4 "Список отличников"

В школе №13 отличников по программированию награждают поездкой в Артек, но счастливых учеников может быть больше, чем свободных мест - ваша задача определить отличников и если их больше чем свободных мест, то отсортировать их по фамилии и вывести тех, кто едет.

Напишите функцию **Artek(s, n)**, которая принимает в качестве параметра строку, содержащую имена отличников и их оценку, и число - количество свободных мест.Функция должна печатать фамилии тех, кто едет.

**Пример:**

**Ввод:**

Сидоров 3, Смирнова 5, Самойкин 4, Емельянов 2, Кац 5, Вдовиченко 5, Северин 4, Субботкин 5

3

**Вывод:**

Вдовиченко, Кац, Смирнова

На вход идут фамилия с оценкой через запятую, а на следующей строке кол-во свободных мести

```python
def Artek(s, n):
    students = s.split(', ')
    students.sort(key=lambda student: student[-1], reverse = True)
    students = [student[:-2] for student in students]
    print(', '.join(sorted(students[:n])))

row = input('Введите строку, содержащую имена отличников и их оценку: ')
n = int(input('Введите количество свободных мест: '))
Artek(row, n)
```

## Задание №5 "Пароль (*)"

Написать программу, которая просит пользователя ввести пароль и подтвердить его (как при обычной регистрации). Пароль должен соответствовать следующим требованиям:

1)В пароле должны быть буквы (Заглавные и прописные)

2)В пароле должны быть цифры

3)В пароле должен быть хотя бы один спец. символ @#$%&*().,-_+

4)Длина пароля не менее 8 символов


**Вывод:**

Если пароль содержит 7 или менее символов - программа
должна печатать "Недопустимый пароль".
Если в пароле содержатся только цифры или только буквы одного
регистра, функция должна печатать "Ненадежный пароль".
Если в пароле содержатся только буквы разных регистров или буквы
одного регистра и цифры, функция должна печатать "Слабый пароль".
Если в пароле содержатся и цифры, буквы разных регистров и спец символы, то
функция должна печатать "Надежный пароль".

Напишит**е функцию passwo**rd_level(), которая
принимает на вход строку и определяет, является ли она надежным паролем.


```python
def password_level(s):
    if len(s) <= 7:
        print('Недопустимый пароль')
    else:
        q = 0
        ss = ''
        for symbol in s:
            if symbol in '@#$%&*().,-_+':
                q = 1
            else:
                ss += symbol  
        if len(ss) != 0:
            if ss.isdecimal():
                if q == 0:
                    print('Ненадежный пароль')
            else:
                if ss.isalpha():
                    if q == 0:
                        if ss.islower() or ss.isupper():
                            print('Ненадежный пароль')
                        else:
                            print('Слабый пароль')
                else:
                    if ss.islower() or ss.isupper():
                        if q == 0:
                            print('Слабый пароль')
                    elif q == 1:
                        print('Надежный пароль')
                    
s = input('Введите пароль: ')
password_level(s)
```
