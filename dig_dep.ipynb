### Задание №2 "Найди кота"

**Формат ввода**

В первой строке записано число n. 
Далее следует n строк.

**Формат вывода**

Напечатайте нужное сообщение в зависимости от того, есть во введенных строчках кот или нет.



Данные сразу анализируются по ходу считывания. Так, если слово содержащее кота найдется, он сразу выдаст МЯУ без необходимости дальнейнего ввода

```python
#Работа через консоль
n = int(input("Введите количество строк: "))
for i in range(n):
    row = input(f'Строка {i+1}: ')
    if row.find('Кот')!= -1 or row.find('кот') != -1:
        print("МЯУ")
        break
    elif i == n-1:
        print("Нет")
 

# Работа через файл


# f = open("test1.dat", encoding="utf-8")
# n = int(f.readline())
# for i in range(n):
#     row = f.readline()
#     if row.find('Кот')!= -1 or row.find('кот') != -1:
#         print("МЯУ")
#         break
#     elif i == n-1:
#         print("Нет")
# f.close()
```

### Задание №3 "Слова и буквы"

**Формат ввода**

Слова разной длины, из которых нужно выбрать кратчайшее и длиннейшее. После этих слов следует слово "стоп".

**Формат вывода**

ДА или НЕТ

```python
word = input("Введите слово: ")
if word == 'стоп':
    print('Нет данных для анализа') #Можно через try except, но мало с этим работал, и не было большой необходимости в этом прежде
else:
    word_min = word
    word_max = word
    while True:
        word = input("Введите слово: ")
        if word == 'стоп':
            break
        if len(word) > len(word_max):
            word_max = word
        elif len(word) < len(word_min):
            word_min = word
    symbols = {}
    for c in word_max:
        if c in symbols.keys():
            symbols[c] += 1
        else:
            symbols[c] = 1
    for k, c in enumerate(word_min):
        if c in symbols.keys():
            if symbols[c] == 0:
                print("НЕТ")
                break
            if k != len(word_min) - 1:
                symbols[c] -= 1    
            else:
                print("ДА")
        else:
            print("НЕТ")
            break
```

Замечание: можно было сделать через list, но dict выглядил интереснее. Через list нужно было бы еще выкидывать символы. Просто пройти по string нельзя, так как слова с повторяющимися символами могут сверяться с словом, где этот символ один. В таком случае ответ был бы ДА. Пример, лото - 2 буквы о, и литография - 1 буква о.

### Задание №4 "Список покупок"

Вы собираетесь в магазин и записываете, что нужно купить.
Напишите программу, которая считывает сначала количество покупок, потом по очереди сами эти покупки, затем выводит их же в том же порядке.


**Формат ввода**

На первой строке - количество покупок.
Затем список покупок, каждая позиция списка на отдельной строке

**Формат вывода**

Тот же самый список

```python
n = int(input("Введите количество покупок"))
list_to_buy = []
for i in range(n):
    list_to_buy.append(input())
print("Список покупок:")
for i in list_to_buy:
    print(i)
```

### Задание №5 "Ххооллоодд"

Очень холодно, вы дрожите, зуб на зуб не попадает, разговариваете соответственно.

**Формат ввода**

Вводится строка.

**Формат вывода**

Выводится та же строка, но каждый символ повторяется дважды

```python
word = input("Введите слово: ")
print(''.join([word[i]*2 for i in range(len(word))]))
```
